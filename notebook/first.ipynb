```python
# 価格を生成する関数
def create_price(M, d) :
    price = np.random.randint(d, 3*d, size=M)
    return price   
```

```python
# alphaを作成する関数
def alpha_star(M) :
    alpha_star = np.random.normal(0, 1, size=M)
    return alpha_star
```

```python
# betaを作成する関数
def beta_star(M, M_prime) :
    beta_star = np.random.normal(0, 1, size=(M, M_prime, 3))
    return beta_star
```

```python
# phi関数
def phi(price ,beta) :
    phi = beta[0]*price + beta[1]*price**2 + beta[2]*(1/price)
    return phi
```

```python
M = 50
# このデータにおけるalpha,betaを作成
alpha=alpha_star(M)
beta=beta_star(M, M)
```

```python

# 価格を入れるリストを作成
price_list = []
# 量を入れるリストを作成
quantity_list = []
for i in range(10000) :
    price = create_price(M, 100)
    price_list.append(price)
    quantity = []
    for m in range(M) :
        sum_beta = 0
        for m_prime in range(M) :
            sum_beta += phi(price[m], beta[m][m_prime])
        quantity.append(alpha[m] + sum_beta + np.random.normal(0, sigma, size=1))
    quantity_list.append(sum(quantity))

```

```python
# 列名の作成
column_names = [f'prodact_{i+1}' for i in range(50)]
```

```python
# データフレームの作成
df_price = pd.DataFrame(price_list, columns=column_names)
```

```python
# 目的関数の定義
def objective_function(params, X, y):
    # paramsは [係数1, 係数2, ..., バイアス] という形
    coef = params[:-1]
    intercept = params[-1]
    
    # 線形回帰モデルの予測値
    y_pred = np.dot(X, coef) + intercept
    
    # 二乗誤差の和を計算
    error = np.sum((y_pred - y) ** 2)
    
    return error
```

```python
# 目的関数の定義（売上の負の値を返す）
def objective_function(prices, coef, intercept):
    # 価格に基づく売上の予測
    sales = np.dot(prices, coef) + intercept
    
    # 売上を最大化するために、負の値を返す（minimizeは最小化を行うため）
    return -sales
```

```python
# 初期価格（例えば、各製品の初期価格を100とする）
initial_prices = np.full(M, 100)
```

```python
print(initial_prices)
```

```python
# 線形回帰モデルの係数とバイアスを取得
coef = lr.coef_
intercept = lr.intercept_

# Nelder-Mead法を用いて最適化
result = minimize(objective_function, initial_prices, args=(coef, intercept), method='Nelder-Mead')

# 最適化された価格
optimal_prices = result.x

print("Optimal Prices:", optimal_prices)
```

```python
optimal_sales = np.dot(optimal_prices, coef) + intercept
print("Optimal Sales:", optimal_sales)
```

```python
def sales_function(price, alpha, beta, phi, sigma):
    M = len(price)
    total_sales = 0
    
    for m in range(M):
        sum_beta = 0
        for m_prime in range(M):
            sum_beta += phi(price[m], beta[m][m_prime])
        
        quantity = alpha[m] + sum_beta + np.random.normal(0, sigma)
        total_sales += quantity
    
    return total_sales
```

```python
def objective_function(price, alpha, beta, sigma):
    # 売上の負の値を返す（売上を最大化するため）
    return -sales_function(price, alpha, beta, phi, sigma)
```

```python
# 最適化の実行
result = minimize(objective_function, initial_prices, args=(alpha, beta, sigma), method='Nelder-Mead')

# 最適化された価格
optimal_prices = result.x

print("Optimal Prices:", optimal_prices)

# 最適化された価格での売上
optimal_sales = -result.fun
print("Optimal Sales:", optimal_sales)
```
